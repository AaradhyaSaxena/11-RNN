```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import json
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage

import glob
import re, os
import pickle
from nltk.stem import PorterStemmer
from gensim.models import TfidfModel
from gensim import similarities
from gensim import corpora
from scipy.cluster import hierarchy
```

```python
def extract_prod_categories(data):
    s = json.loads(data)
    s = s[0]
    s = s.replace(" >> ", "|")
    
    return s;

def extract_prod_slot(s):
    arr = []
    for i in range(len(s)):
        if s[i] == '|':
            arr.append(i)
    
    last = len(arr)
    if last > 0:
        return s[:arr[0]]
    else:
        return s;
```

```python
df = pd.read_csv("dataset.csv")
```

```python
# make new column >> product sub-categories
df['product_category'] = df['product_category_tree'].apply(extract_prod_categories)
# make new column >> product prime category
df['slot'] = df['product_category'].apply(extract_prod_slot)

df = df.drop(['product_category_tree'], axis=1)
df.head(3)
```
